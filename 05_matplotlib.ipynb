```python
import matplotlib.pyplot as plt

import numpy as np
np.set_printoptions(precision=3 , suppress= True) # this is just to make the output look better
%matplotlib inline
```

```python
x = np.linspace(0,10,20)
y = x ** 2
```

```python
xx = np.linspace(0,2*np.pi,100)
psin = np.sin(xx)
pcos = np.cos(xx)
```

```python
plt.plot(xx,psin, linewidth=5)
plt.plot(xx,pcos)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y, 'r--o', x, y ** 1.1, 'bs', x, y ** 1.2, 'g^-' );
plt.xlabel('x', size=15)
plt.ylabel('y', size=15)
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
plt.legend(['red','blue', 'green'], fontsize=16)
plt.title('the plot', fontsize=16);
```

```python
plt.figure(figsize=(15,5))
plt.xlabel('days', size=15)
plt.ylabel('temperature, $^\circ$C', size=15)
plt.plot(temp[:1000,3]/10)
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
plt.legend(['Hamburg temperature'], fontsize=16)
plt.title('The plot is created by John Doe', fontsize=16);
```

```python
def plot_temp(xmin=0, xmax=1000, lw=2, color='blue'):
    plt.figure(figsize=(15,5))
    plt.xlabel('days', size=15)
    plt.ylabel('temperature, $^\circ$C', size=15)
    plt.plot(temp[xmin:xmax,3]/10, lw=lw, color=color)
    plt.xticks(fontsize=16);
    plt.yticks(fontsize=16);
    plt.legend(['Hamburg temperature'], fontsize=16)
    plt.title('The plot is created by John Doe', fontsize=16);
```

```python
plot_temp(lw=1, color='red')
```
