```python
Select and plot only data for October
Calculate monthly means for years from 1990 to 1999 and plot them
```

```python
plt.plot(temp[(temp[:,1]==6)][:,3]/10)
```

```python
temp[:,0] < 1999
```

```python
subset = temp[(1990 <= temp[:,0]) & (temp[:,0] <= 1999)]
```

```python
mm = []
for year in range(1990, 2000):
    for mon in range(1,13):
        mon_data = subset[(subset[:,0]==year) & (subset[:,1]==mon)][:,3]/10
        mon_data_mean = mon_data.mean()
        mm.append(mon_data_mean)
```

```python
plt.plot(mm)
```
